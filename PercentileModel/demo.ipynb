```python
import matplotlib.pyplot as plt
```

```python
totalCnts = [10,100,1000,10000]
regressCnts = [0,2,8,20]
texts_regressCnt = ["regressCnt="+str(x) for x in regressCnts]
regressRates = [regressCnts[i]/totalCnts[i] for i in range(len(regressCnts))]
```

```python
texts_percentile = ['worse than 0%', 'worse than 35%', 'worse than 75%', 'worse than 60%']
```

```python
texts = texts_percentile#texts_regressCnt#texts_percentile

plt.scatter(totalCnts, regressRates)
plt.plot(totalCnts, regressRates)
for i in range(len(regressCnts)):
    plt.text(totalCnts[i], regressRates[i], texts[i])
plt.xlabel("Deploy Node Counts")
plt.ylabel("Regress Rate")
# plt.show()
```
